## IMPORT LIBRARY

```python
# Import library
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
path1 = '/content/drive/MyDrive/Colab Notebooks/Heart Attack Analysis & Prediction Dataset/heart.csv'
path2 = '/content/drive/MyDrive/Colab Notebooks/Heart Attack Analysis & Prediction Dataset/o2Saturation.csv'

df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2,names=["o2Saturation"])
```

## PRE-PROCESSING

```python
df = df1.merge(df2, left_index=True, right_index=True)
df.head(8)
```

```python
df.rename(columns={"cp": "chest_pain",
                   "trtbps":"blood_pressure",
                   "chol":"cholestoral",
                   "fbs":"fasting_blood_sugar",
                   "thalachh":"max_heart_rate",
                   "exng":"angina",
                   "caa": "n_vessels",
                   "output":"heart_attack"},
          inplace=True)

df['sex'].replace({0:'Female', 1:'Male'}, inplace=True)

li = df.columns.tolist()
li[0],li[1],li[-2],li[-1] = li[1],li[0],li[-1],li[-2]
df = df[li]

df.head()
```

## MODELLING

### Using Logistic Regression

```python
# Memilih fitur yang akan digunakan sebagai variabel independen (X) dan target (y)
X = df1.iloc[:, :-1].values
y = df1.iloc[:, -1].values

# Membagi data menjadi data pelatihan dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=65)

# Inisialisasi model logistic regression
model = LogisticRegression()

# Melatih model pada data pelatihan
model.fit(X_train, y_train)

# Melakukan prediksi pada data uji
y_pred = model.predict(X_test)

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)

# Menampilkan hasil akurasi
print("Akurasi: ", accuracy)
```
